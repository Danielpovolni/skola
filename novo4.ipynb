```python
import matplotlib.pyplot as plt
import pandas as pd 

faks=[
    ["Marko", 6,6,6,6,9,9],
    ["Vesna", 6,7,6,7,9,9],
    ["Sandra", 7,7,6,7,9,9],
    ["Ivona", 6,7,6,7,8,9],
    ["Sara", 9,10,7,9,10,10]
     ]
predmeti = pd.DataFrame(faks)
predmeti.columns = ["Ime","IKT","Menadzment", "Matematika", "Rac_Sistemi","Ekonomija","Architektura"]
ocene = predmeti.set_index("Ime")
ocene
```

```python
frekv = ocene["Ekonomija"].value_counts()
print("Ocene iz ekonomije: ", frekv.index)
print("frekvencije iz ekonomije: ", frekv.values)
frekv
```

```python
plt.figure(figsize=(5,5))
plt.pie(frekv.values, labels=frekv.index)
plt.title("ucestalost ocena")
plt.show()
plt.close()
```

```python
faks=[
    ["Marko", 6,6,6,6,9,9],
    ["Vesna", 6,7,6,7,9,9],
    ["Sandra", 7,7,6,7,9,9],
    ["Ivona", 6,7,6,7,8,9],
    ["Sara", 9,10,7,9,10,10]
     ]
predmeti = pd.DataFrame(faks)
predmeti.columns = ["Ime","IKT","Menadzment", "Matematika", "Rac_Sistemi","Ekonomija","Architektura"]
ocene = predmeti.set_index("Ime")
ocene
for student in ocene.index:
    print(student, "Ima prosecnu ocenu ", round(ocene.loc[student].mean(),2))
```

```python
ocene["Prosek_studenata"] = 0.0
for student in ocene.index:
    ocene.loc[student, "Prosek_studenata"] = ocene.loc[student, "IKT":"Architektura"].mean()
ocene
```

```python
ocene["Prosek_predmeti"] = 0.0
for predmet in ocene.columns:
    ocene.loc["Prosek_predmeti", predmet] = ocene.loc["Marko":"Sara", predmet].mean()
ocene
```
